## Exercise 1

**Follow along with these steps**
* Create a figure object called fig using plt.figure()
* Use add_axes to add an axis to the figure canvas at [0,0,1,1]. Call this new axis ax.
* Plot (x,y) on that axes and set the labels and titles to match the plot below:

```python
fig1=plt.figure()
x = np.arange(0,100)
y = x*2
ax=fig1.add_axes([0,0,1,1])
ax.plot(x,y,color='blue')
plt.xlabel("x")
plt.ylabel("y")
plt.title("title")

plt.show()
```

```python
fig=plt.figure()

x1 = np.arange(0,100)
z = (x1)**2
x2=np.arange(20.0,22.5,0.5)
y=x2*2


ax1=fig.add_axes([0,0,1,1])
ax1.set_ylim(0,10000)
ax1.set_xlim(0,100)
ax1.plot(x1,z,color='blue')
ax1.set_xlabel("X")
ax1.set_ylabel("Z")

ax2=fig.add_axes([0.2,0.5,.4,.4])
ax2.set_ylim(30,50)
ax2.set_xlim(20,22)
ax2.plot(x2,y,color='blue')
ax2.set_xlabel("X")
ax2.set_ylabel("Y")
ax2.set_title("zoom")

plt.show()
```

**Now plot (x,y) and (x,z) on the axes. Play around with the linewidth and style**

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0,100)
y = x*2
plt.figure()
plt.subplot(1,2,1)
plt.plot(x,y,linestyle='dotted',color='violet',linewidth=3)

z = x**2
plt.subplot(1,2,2)
plt.plot(x,z,color='green',linestyle='dashed',linewidth=2)

plt.show()
```

**See if you can resize the plot by adding the figsize() argument in plt.subplots() are copying and pasting your previous code.**

```python
x = np.arange(0,100)
y = x*2
plt.figure(figsize=(12,2))
plt.subplot(1,2,1)
plt.plot(x,y,color='red',linewidth='2',linestyle='dashed')


z = x**2
plt.subplot(1,2,2)
plt.plot(x,z,color='green',linestyle='dotted',linewidth='3')

plt.show()
```
